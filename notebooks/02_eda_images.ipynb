# EDA — FGVC-Aircraft Images
Class balance, sample grid, image sizes.

```python
from torchvision.datasets import FGVCAircraft
from collections import Counter
import matplotlib.pyplot as plt
ds = FGVCAircraft(root='../data/raw/fgvc_aircraft', split='train', annotation_level='variant', download=False)
print('classes:', len(ds.classes), 'samples:', len(ds))
```

```python
counts = Counter(label for _, label in ds)
plt.hist(list(counts.values()), bins=30); plt.title('Per-class image counts'); plt.show()
```

```python
fig, ax = plt.subplots(3, 4, figsize=(12, 8))
for a, (img, lbl) in zip(ax.flatten(), ds): a.imshow(img); a.set_title(ds.classes[lbl], fontsize=8); a.axis('off')
plt.tight_layout(); plt.show()
```

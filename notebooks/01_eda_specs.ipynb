# EDA — Aircraft Specs

Explores `data/processed/aircraft_specs.csv`: distributions of range, MTOW, manufacturer counts, missingness.

```python
import pandas as pd, matplotlib.pyplot as plt, seaborn as sns
df = pd.read_csv('../data/processed/aircraft_specs.csv')
df.info()
```

```python
df['range_km'].hist(bins=30); plt.title('Range distribution (km)'); plt.show()
```

```python
df['manufacturer'].value_counts().head(15).plot(kind='barh'); plt.title('Manufacturers'); plt.show()
```

```python
df.isna().mean().sort_values(ascending=False).plot(kind='barh'); plt.title('Missingness'); plt.show()
```
